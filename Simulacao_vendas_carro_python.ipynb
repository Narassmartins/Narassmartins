```python

nome = input("Digite o nome do cliente: ")
carro = input("Digite o modelo do carro: ")
valor_carro = float(input("Digite o valor do carro: "))
entrada = float(input("Digite o valor da entrada: "))
parcelas = int(input("Digite a quantidade de parcelas: "))


valor_financiado = valor_carro - entrada
valor_parcela = valor_financiado / parcelas

print("\n--- SIMULAÇÃO DE VENDA ---")
print(f"Cliente: {nome}")
print(f"Carro: {carro}")
print(f"Valor do carro: R$ {valor_carro:.2f}")
print(f"Entrada: R$ {entrada:.2f}")
print(f"Valor financiado: R$ {valor_financiado:.2f}")
print(f"Parcelas: {parcelas}x de R$ {valor_parcela:.2f}")


if valor_parcela <= 1500:
    print("Parcela acessível para o cliente ")
else:
    print("Parcela alta, precisa negociar ")

if entrada >= valor_carro * 0.2:
    print("Entrada boa, aumenta chance de aprovação ")
else:
    print("Entrada baixa, pode dificultar aprovação ")


print("\nEvolução do pagamento:")
total_pago = entrada

for i in range(1, parcelas + 1):
    total_pago += valor_parcela
    print(f"Após pagar {i} parcelas: R$ {total_pago:.2f}")


print("\nSimulando pagamento até quitar o carro:")
saldo = valor_financiado
contador = 0

while saldo > 0:
    saldo -= valor_parcela
    contador += 1

print(f"O cliente quitou o carro em {contador} parcelas.")
```
